## Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pn
```

```python
import numpy as ny
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,4,9,16,25,36,49,64,81]

plt.plot(x,y,color='r',linestyle='-.',linewidth=2.0)
plt.xlabel('Numbers')
plt.ylabel('Square of numbers')
plt.title('My first plot')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,4,9,16,25,36,49,64,81]

plt.scatter(x,y,color='b',linewidth=2.0)
plt.show()
plt.plot(x,y,color='r',linestyle='-.',linewidth=2.0)
plt.xlabel('Numbers')
plt.ylabel('Square of numbers')
plt.title('My first plot')
plt.grid()
plt.show()
```

```python
x=ny.arange(-50,50)
y=x**2
plt.plot(x,y,color='r',linestyle='-.',linewidth=2.0)
plt.plot(x,2**x)
plt.xlabel('X')
plt.ylabel('Y=X**2')
plt.title('Parabolic plot')
plt.grid()
plt.show()
```

```python
x=[1,4,6,4]
y=[1,2,3,4]
plt.plot(x,y,color='y',linewidth=3.0,linestyle='-.')
plt.xlabel("numbers")
plt.ylabel("indices")
plt.title("Myplot")
plt.show()
```

```python
plt.plot([1,6,4,3],[7,0,1,2],'or')
plt.grid()
plt.show()
```

```python
t=ny.arange(0.1,5.,2.6)
plt.plot(t,t**2,'rs',label='2^')
plt.plot(t,t**2.2,'g^',label='2.2^')
plt.plot(t,t**2.5,'b',label='2.5^')
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[4,6,8,9]
line=plt.plot(x1,y1,x2,y2)
plt.setp(line[0],color='r',linewidth=3.0,linestyle='-')
plt.setp(line[1],color='g',linewidth=2.5)
plt.show(line)
```
