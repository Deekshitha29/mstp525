```python
import pandas as pd
```

```python
drug=pd.read_csv('drug.csv')
drug
```

```python
drug.isnull().sum()
```

```python
drug.duplicated().sum()
```

```python
drug.drop_duplicates()
```

```python
drug['Drug'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
drug['Sex']=lb.fit_transform(drug['Sex'])
drug['Cholesterol']=lb.fit_transform(drug['Cholesterol'])
drug['Drug']=lb.fit_transform(drug['Drug'])
drug['BP']=lb.fit_transform(drug['BP'])
drug.head()
```

```python
Y=drug['Drug']
X=drug.drop('Drug',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,Y_train)
```

```python
Y_pred=tree.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
con=confusion_matrix(Y_pred,Y_test)
con
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred,Y_test)
```
